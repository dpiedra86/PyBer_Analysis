```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
fig, ax=plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label='Chicago')
plt.gca().invert_yaxis()
plt.legend()
plt.xlabel("Fare($)")
plt.ylabel("Date")
```

```python
plt.scatter(y_axis, x_axis, c=y_axis, s=y_axis, label="Chicago")
plt.legend()
plt.xlabel("Fare($)")
plt.ylabel("Date")

```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, 
           
           s=[i * 5 for i in y_axis], 
           
           c="lime", 
           
           linewidths=2, 
           
           edgecolors="k", 
           
           alpha=.2,
           
           label="Boston")

plt.legend()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.gca().invert_yaxis()
```

```python
explode = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["olivedrab","greenyellow","yellowgreen","lawngreen",
          "limegreen","lime","seagreen","mediumseagreen",
          "springgreen","mediumspringgreen","mediumaquamarine","aquamarine"]

plt.subplots(figsize=(11,11))
plt.pie(y_axis, labels=x_axis, 
        explode=explode, 
        autopct="%.1f%%",
        colors = colors
       frame = true)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.pie(y_axis,labels=x_axis,
      explode = explode,
      colors = colors,
      shadow = 0,
      startangle = 260,
      autopct="%.1f%%",
      textprops = {"color" : "blue"},
      )
plt.gca().invert_yaxis()
plt.show()
```
